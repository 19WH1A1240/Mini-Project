# Changing object types to categorical values

```python
finaldata['Base']=finaldata.Base.astype('category')
finaldata['Dispatching_base_num']=finaldata.Dispatching_base_num.astype('category')
finaldata['Pickup_date']=finaldata.Pickup_date.astype('category')
finaldata['Affiliated_base_num']=finaldata.Affiliated_base_num.astype('category')
finaldata['dispatching_base_number']=finaldata.dispatching_base_number.astype('category')
finaldata['date']=finaldata.date.astype('category')
finaldata.info()
```

# Using mean for float,int and bfill/ffill for categorical to fill the Null values

```python
finaldata=finaldata.copy()
finaldata["Lat"].fillna(finaldata["Lat"].mean(), inplace=True)
finaldata["Lon"].fillna(finaldata["Lon"].mean(), inplace=True)
finaldata["locationID"].fillna(finaldata["locationID"].mean(), inplace=True)
finaldata["active_vehicles"].fillna(finaldata["active_vehicles"].mean(), inplace=True)
finaldata["trips"].fillna(finaldata["trips"].mean(), inplace=True)
finaldata["DayOfWeekNum"].fillna(finaldata["DayOfWeekNum"].mean(), inplace=True)
finaldata["MonthDayNum"].fillna(finaldata["MonthDayNum"].mean(), inplace=True)
finaldata["HourOfDay"].fillna(finaldata["HourOfDay"].mean(), inplace=True)
finaldata["MinOfDay"].fillna(finaldata["MinOfDay"].mean(), inplace=True)
```

```python
finaldata['Base'] = finaldata['Base'].fillna(method='ffill')
finaldata['Dispatching_base_num'] = finaldata['Dispatching_base_num'].fillna(method='bfill')
finaldata['Pickup_date'] = finaldata['Pickup_date'].fillna(method='bfill')
finaldata['Affiliated_base_num'] = finaldata['Affiliated_base_num'].fillna(method='bfill')
finaldata['dispatching_base_number'] = finaldata['dispatching_base_number'].fillna(method='bfill')
finaldata['date'] = finaldata['date'].fillna(method='bfill')
finaldata['Date/Time']=finaldata['Date/Time'].fillna(method='ffill')
```

```python
dfclusters = df[['Lat','Lon']]
dfclusters.dtypes
```

```python
from sklearn.cluster import KMeans
inertia = []
K = range(1,10)
for k in K:
    km = KMeans(n_clusters=k,max_iter=50)
    km.fit(dfclusters)
    inertia.append(km.inertia_)
```

```python
plt.figure(figsize=(8,4))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel
('inertia')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
km = KMeans(n_clusters=6, max_iter = 300, random_state = 12345)
km.fit(dfclusters)
```

```python
import tkinter as tk
from tkinter import filedialog
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

root= tk.Tk()

canvas1 = tk.Canvas(root, width = 400, height = 400,  relief = 'raised')
canvas1.pack()

label1 = tk.Label(root, text='UBER DATA ANALYSIS')
label1.config(font=('helvetica', 20))
canvas1.create_window(190, 20, window=label1)

label2 = tk.Label(root, text='Type Number of Clusters:')
label2.config(font=('helvetica', 12))
canvas1.create_window(200, 165, window=label2)

# label2 = tk.Label(root, text='Enter the latitude:')
# label2.config(font=('helvetica', 12))
# canvas1.create_window(200, 165, window=label2)

# entry2 = tk.Entry (root) 
# canvas1.create_window(190, 185, window=entry2)

# label4 = tk.Label(root, text='Enter the Longitude:')
# label4.config(font=('helvetica', 12))
# canvas1.create_window(200, 365, window=label2)

entry1 = tk.Entry (root) 
canvas1.create_window(190, 185, window=entry1)


def getExcel ():
    
    global df
    import_file_path = filedialog.askopenfilename()
    read_file = pd.read_excel (import_file_path)
    df = DataFrame(read_file,columns=['Lat','Lon'])  
    
browseButtonExcel = tk.Button(root,text=" Import Excel File ", command=getExcel, bg='green', fg='white', font=('helvetica', 15, 'bold'))
canvas1.create_window(200, 80, window=browseButtonExcel)

def getKMeans ():
    global df
    global numberOfClusters
    numberOfClusters = int(entry1.get())
    kmeans = KMeans(n_clusters=numberOfClusters).fit(df)
    centroids = kmeans.cluster_centers_
    label3 = tk.Label(root, text= centroids)
    canvas1.create_window(200, 370, window=label3)
    figure1 = plt.Figure(figsize=(4,3), dpi=100)
    ax1 = figure1.add_subplot(111)
    ax1.scatter(df['Lat'], df['Lon'], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
    ax1.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
    ax1.get_legend()
    scatter1 = FigureCanvasTkAgg(figure1, root) 
    scatter1.get_tk_widget().pack(side=tk.RIGHT, fill=tk.BOTH)
    
processButton = tk.Button(root,text=' Process k-Means ', command=getKMeans, bg='brown', fg='white', font=('helvetica', 15, 'bold'))
canvas1.create_window(200, 250, window=processButton)

root.mainloop()
```
